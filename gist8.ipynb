```python
from MSEIR import MSEIR

city_kwargs = {'n_periods': 200, 'N': 100000,
               'R_0': 2.0, 'tLat': 2, 'tInf': 2, 'tHosp': 2,
               'pMild': 0.90, 'pFatal': 0.20, 'pOvf':0.20,
               'Q': 200, 'S0': 100000-1,
               'E0': 0, 'I0': 1, 'H0': 0, 'R0': 0, 'D0': 0}

m = MSEIR(**city_kwargs)
res = m.solve(optimize=True, solver='SLSQP', bounds=(0,0.9), freq=10, hor=10)
m.plot(res[::10], comps='SEIHRD')
```
